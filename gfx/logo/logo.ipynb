```python
import os
import matplotlib.pyplot as plt
import ska
```

# Load data

```python
sun = ska.Spectrum(os.path.join("..", "data", "hst_sun.csv") )
```

```python
J = ska.Filter("2MASS/2MASS.J")
H = ska.Filter("Paranal/VISTA.H")
K = ska.Filter("Paranal/VISTA.Ks")
```

# Blackbody

```python
fig, ax = plt.subplots()

ax.plot(sun.Wavelength, sun.Flux, color="black")

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlim(0.15, 5)
ax.set_ylim(bottom=0.1)
plt.axis("off")
fig.savefig("logo_sun_black.png", dpi=300, transparent=True)
```

```python
fig, ax = plt.subplots()

ax.plot(sun.Wavelength, sun.Flux, color="white")

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlim(0.15, 5)
ax.set_ylim(bottom=0.1)
plt.axis("off")
fig.savefig("logo_sun_white.png", dpi=300, transparent=True)
```

# Filters

```python
fig, ax = plt.subplots()

for f, fn in zip([J, H, K], ["J", "H", "Ks"]):

    ax.plot(f.wave, f.trans)
    ax.fill_between(f.wave, 0, f.trans, alpha=0.2)

ax.set_xlim(1, 2.5)
ax.set_ylim(bottom=0)
plt.axis("off")

fig.savefig(f"logo_filters.png", dpi=300, transparent=True)
```

```python

i=0
for f, fn in zip([J, H, K], ["J", "H", "Ks"]):
    fig, ax = plt.subplots()

    ax.plot(f.wave, f.trans, color=f'C{i:d}')
    ax.fill_between(f.wave, 0, f.trans, alpha=0.2, color=f'C{i:d}')

    ax.set_xlim(1, 2.5)
    ax.set_ylim(bottom=0)
    plt.axis("off")

    fig.savefig(f"logo_filter_{fn}.png", dpi=300, transparent=True)
    i = i+1
```

```python
fig, ax = plt.subplots()


ax.plot(sun.Wavelength, sun.Flux, label="Sun")

ax.plot(H.wave, H.trans, label="H")
ax.plot(K.wave, K.trans, label="H")

ax.set_xlabel("Wavelength (nm)")
ax.set_ylabel("Flux (W/m^2/nm)")
ax.set_title("Solar Spectrum")
ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
```
